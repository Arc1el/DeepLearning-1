 - 가설함수 $f = w_1*x_1 + w_2*x_2 + b$

```python
import numpy as np
```

```python
x_data = np.array([[1.0, 2.0], 
                   [2.0, 4.0], 
                   [3.0, 9.0]])
```

```python
W = np.random.rand(1,2)
b = np.random.rand(1)
```

```python
def func1(X):
  return np.dot(W, X.transpose()) + b
```

```python
def numerical_diff(f, X):
  h = 1e-4
  grad = np.zeros_like(X)
  print(grad)

  vec1_dim = X.shape[0], vec2_dim = X.shape[1] 
  for x_idx in range(vec1_dim):
    for y_idx in range(vec2_dim):
      temp_value = X[x_idx, y_idx]
      X[x_idx, y_idx] = temp_value + h
      f_f = f(X)

      X[x_idx, y_idx] = temp_value - h
      f_b = f(X)

      grad[x_idx, y_idx] = (f_f - f_b) / (2*h)
      X[x_idx, y_idx] = temp_value

  return grad
```

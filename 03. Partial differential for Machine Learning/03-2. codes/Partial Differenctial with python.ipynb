# 딥러닝을 위한 미분 구현하기

## 1. Simple 미분 계산기 구현하기
 - 입력변수가 한개인 함수에 대한 미분값을 자동으로 구하는 프로그램
 - input: flot형 숫자 x
 - output: 함수 f(x)에 대한 미분 결과 
 - f(x): x의 제곱을 리턴하는 함수

```python
import numpy as np
```

```python
def func1(x):
  return x**2
```

```python
def simple_numerical_diff(f, x):
  h = 10e-5

  diff = (f(x + h) - f(x - h)) / (2*h)
  return diff
```

```python
input_data = 4.0
my_func = func1
simple_numerical_diff(my_func, input_data)
```

## 2. gradient 계산기 구현하기
 - 입력변수가 한개이상인 함수에 대한 미분값을 자동으로 구하는 프로그램
 - input: flot형 숫자 한 개 이상인 numpy array자료형의 벡터 X
 - output: 함수 f(X)에 대한 미분 결과 벡터 
 - f(X): 2x+3xy+y^3을 리턴하는 함수

```python
input_weight = [1.0, 2.0]
input_data = np.array(input_weight)
```

```python
def func2(X):
  x = X[0]
  y = X[1]
  return 2*x + 3*x*y + np.power(y, 3)
```

```python
func2(input_data)
```

```python
def numerical_diff(f, X):
  h = 1e-4
  grad = np.zeros_like(X)

  for idx in range(X.size):
    temp_value = X[idx]
    X[idx] = temp_value + h
    f_f = f(X)

    X[idx] = temp_value - h
    f_b = f(X)

    grad[idx] = (f_f - f_b) / (2*h)
    X[idx] = temp_value

  return grad

```

```python
f2 = func2

numerical_diff(f2, input_data)
```
